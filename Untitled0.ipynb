```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

def preprocess_ckd_dataset(input_path, output_path):
    # Load the dataset
    df = pd.read_csv(input_path)

    # Drop irrelevant or inconsistent columns if present
    drop_cols = ['id', 'wc', 'rc']
    df.drop(columns=[col for col in drop_cols if col in df.columns], inplace=True)

    # Drop rows where target label is missing
    df.dropna(subset=['classification'], inplace=True)

    # Map the classification column to binary labels
    df['classification'] = df['classification'].str.lower().str.strip().replace({'ckd\t': 'ckd'})
    df['label'] = df['classification'].map({'ckd': 1, 'notckd': 0})
    df.drop(columns=['classification'], inplace=True)

    # Handle and encode categorical columns
    label_encoders = {}
    for col in df.select_dtypes(include='object').columns:
        df[col] = df[col].astype(str).str.lower().str.strip()  # Clean strings
        le = LabelEncoder()
        df[col] = le.fit_transform(df[col])
        label_encoders[col] = le

    # Drop remaining rows with missing values
    df.dropna(inplace=True)

    # Save the cleaned dataset
    df.to_csv(output_path, index=False)
    print(f"Preprocessed dataset saved to: {output_path}")
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import accuracy_score, classification_report
from xgboost import XGBClassifier

# Load CKD dataset
file_path = '/content/drive/MyDrive/ckd1/processed_kidney_disease.csv'
df = pd.read_csv(file_path)

# Preprocessing
# Step 1: Handle missing values (can use median or mode imputation)
for col in df.columns:
    if df[col].dtype == 'object':
        df[col].fillna(df[col].mode()[0], inplace=True)
    else:
        df[col].fillna(df[col].median(), inplace=True)

# Step 2: Encode categorical features
label_encoders = {}
for col in df.select_dtypes(include='object').columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le

# Step 3: Split features and target
# Change 'classification' to the actual name of your target column
X = df.drop('classification', axis=1)  # Changed 'class' to 'classification'
y = df['classification']  # Changed 'class' to 'classification'

# Optional: Normalize or standardize numerical features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 4: Train-test split with stratification
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42, stratify=y)

# Step 5: Calculate class imbalance ratio
scale_pos_weight_value = (y_train == 0).sum() / (y_train == 1).sum()

# Step 6: Train XGBoost model
model = XGBClassifier(eval_metric='logloss', use_label_encoder=False, scale_pos_weight=scale_pos_weight_value, random_state=42)
model.fit(X_train, y_train)

# Step 7: Predict and evaluate
y_pred = model.predict(X_test)

print("âœ… Accuracy:", round(accuracy_score(y_test, y_pred) * 100, 2), "%")
print("\nðŸ“Š Classification Report:\n", classification_report(y_test, y_pred, target_names=["notckd", "ckd"])) # You might need to adjust target_names
```

```python
import pandas as pd
import numpy as np
from xgboost import XGBClassifier
from sklearn.preprocessing import LabelEncoder

class CKDPredictor:
    def __init__(self):
        self.model = None
        self.label_encoders = {}
        self.feature_columns = []

    def load_and_train(self, path):
        df = pd.read_csv(path)

        # Drop rows without target
        df.dropna(subset=['classification'], inplace=True)
        df['label'] = df['classification'].map({'ckd': 1, 'notckd': 0})
        df.drop(columns=['classification'], inplace=True)

        # Encode categorical features
        for col in df.select_dtypes(include='object').columns:
            le = LabelEncoder()
            df[col] = le.fit_transform(df[col].astype(str))
            self.label_encoders[col] = le

        df.dropna(inplace=True)

        X = df.drop(columns=['label'])
        y = df['label']
        self.feature_columns = X.columns.tolist()

        self.model = XGBClassifier(eval_metric='logloss', use_label_encoder=False)
        self.model.fit(X, y)

    def predict(self, input_dict):
        if self.model is None:
            raise Exception("Model not trained. Run load_and_train() first.")

        input_df = pd.DataFrame([input_dict])

        # Encode categorical columns using saved label encoders
        for col, le in self.label_encoders.items():
            if col in input_df.columns:
                # Check if the value is already numeric
                if not pd.api.types.is_numeric_dtype(input_df[col]):
                    try:
                        # Transform the entire column if it's categorical
                        input_df[col] = le.transform(input_df[col].astype(str))
                    except ValueError as e:
                        # Handle unknown values gracefully
                        print(f"Warning: Value '{input_df[col].values[0]}' for column '{col}' not found in training data. Using 0.")
                        input_df[col] = 0  # Assign a default value
                #else:
                    #pass  # If the value is already numeric, no need to transform #No need for this else block
            else:
                # If column is not in input, add it with a default value
                input_df[col] = 0

        # Ensure all expected features are present and in the correct order
        input_df = input_df[self.feature_columns]

        # Ensure data types are numeric (float or int) - this is likely needed by XGBoost
        # The issue was trying to convert object types with mixed int and float to float would fail
        # so convert to numeric before dropping
        for col in input_df.select_dtypes(include=['number']):
            input_df[col] = pd.to_numeric(input_df[col])


        pred = self.model.predict(input_df)[0]
        return "CKD" if pred == 1 else "Not CKD"
```
